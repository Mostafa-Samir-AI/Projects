# Dropping repeated columns

```python
kpd = kp_data.drop(columns={"clv"})
```

# Scaling

```python
from sklearn.preprocessing import MinMaxScaler , StandardScaler
```

```python
scale = StandardScaler()
```

```python
kpd.loc[:,["minutes_watched" , "clv_wins"]] = scale.fit_transform(kpd.loc[:,["minutes_watched" , "clv_wins"]])
```

```python
kpd
```

```python
kpd.drop(columns={"Unnamed: 0"} , inplace=True)
kpd
```

# modeling

```python
# deciding the categorical columns
categorical_columns = [2,3]  
```

```python
from kmodes.kprototypes import KPrototypes
```

```python
for k in range(2,24):
    kproto = KPrototypes(n_clusters=k, init='Huang', verbose=2)

    clusters = kproto.fit_predict(kpd, categorical=categorical_columns)


    print("Labels:", clusters)
    print("Centroids:", kproto.cluster_centroids_)
```
