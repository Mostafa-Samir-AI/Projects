## look for Artifact in Mlflow

```python
import mlflow
from mlflow.tracking import MlflowClient 
```

```python
mlflow.set_tracking_uri("http://localhost:5000")
client = MlflowClient()
```

```python
experiment_id = current_exp["experiment_id"]
experiment_id
```

```python
# retreving model info , when k = 10
my_model_info = dict(mlflow.search_runs([experiment_id] , order_by=["params.K"]).loc[0,:])
```
